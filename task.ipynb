```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = "https://www.indiatvnews.com/business/news-india-gdp-quickest-to-fall-fastest-to-recover-673985"
try:
    page = urlopen(url)
except:
    print("Error opening the URL")
```

```python
soup = BeautifulSoup(page, 'html.parser')
```

```python
content = soup.find('div', {"class":"content"})
```

```python
article = ''
for i in content.findAll('p'):
    article = article + ' ' +  i.text
```

```python
with open('scraped_text.txt', 'w') as file:
    file.write(article)
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
from nltk.chunk import ne_chunk
```

```python
content = open("scraped_text.txt", "r").read()
print(content)
```

```python
ne_tree = ne_chunk(pos_tag(word_tokenize(content)))
```

```python
print(ne_tree)
```

```python
def preprocess(sent):
    sent = nltk.word_tokenize(sent)
    sent = nltk.pos_tag(sent)
    return sent
```

```python
sent = preprocess(content)
sent
```

```python
pattern = 'NP: {<DT>?<JJ>*<NN>}'
```

```python
cp = nltk.RegexpParser(pattern)
cs = cp.parse(sent)
print(cs)
```

```python
from nltk.chunk import conlltags2tree, tree2conlltags
from pprint import pprint

iob_tagged = tree2conlltags(cs)
pprint(iob_tagged)
```

```python
from bs4 import BeautifulSoup
import requests
import re
import en_core_web_sm
nlp = en_core_web_sm.load()
import spacy
from spacy import displacy
from collections import Counter
```

```python
def url_to_string(url):
    res = requests.get(url)
    html = res.text
    soup = BeautifulSoup(html, 'html5lib')
    for script in soup(["script", "style", 'aside']):
        script.extract()
    return " ".join(re.split(r'[\n\t]+', soup.get_text()))
```

```python
ny_bb = url_to_string('https://www.indiatvnews.com/business/news-india-gdp-quickest-to-fall-fastest-to-recover-673985')
article = nlp(ny_bb)
len(article.ents)
```

```python
labels = [x.label_ for x in article.ents]
Counter(labels)
```

```python
items = [x.text for x in article.ents]
Counter(items).most_common(3)
```

```python
sentences = [x for x in article.sents]
print(sentences[20])
```

```python
displacy.render(nlp(str(sentences[20])), jupyter=True, style='ent')
```
